```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset from sklearn
iris = load_iris()
iris_data = pd.DataFrame(iris.data, columns=iris.feature_names)

# Display the first few rows of the dataset to get an overview
iris_data.head()
# Summary statistics
print(iris_data.describe())

# Visualize the distribution of sepal lengths for different species
plt.figure(figsize=(8, 6))
sns.histplot(data=iris_data, x='sepal length (cm)', bins=20, kde=True)
plt.title('Distribution of Sepal Lengths')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Count')
plt.show()

# Explore the relationships between sepal length and sepal width for different species
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris_data, x='sepal length (cm)', y='sepal width (cm)', hue=iris.target)
plt.title('Relationship between Sepal Length and Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()






```
